1. Bacalah file `avocado.json` dari folder `data_tugas3` kedalam variabel `avocado`. Lalu urutkan data avocado terhadap tanggal, pada method sort tersebut tambahkan argument `ignore_index=True`. Simpan hasil akhir dalam variabel `dataframe_1`

```python
import pandas as pd
import numpy as np
```

```python
avocado = pd.read_json('data_tugas3/avocado.json', orient='split')
avocado = avocado.drop('Unnamed: 0',axis=1)
dataframe_1 =  avocado.sort_values(by=['Date'], ignore_index=True)
dataframe_1
```
