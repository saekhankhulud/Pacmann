_Load_ `diabetes.csv`

```python
# Jawaban anda di sini
import pandas as pd
import numpy as np
df = pd.read_csv('diabetes.csv')
df.head()
```

1. Apakah ada kolom yang 40% dari datanya adalah _missing value_? Bila ada, hapus dari tabel! 

```python
# Jawaban anda di sini
df_2 = df.copy()
df_2.drop(df_2.columns[(df.isna().sum()/df.isna().count()>0.4)], axis=1, inplace=True)
df_2.head()
```

2. Pada kolom `class`, temukan nilai-nilai uniknya. Ubah semua nilainya menjadi "Diabetes" (untuk yang positif) atau "Not diabetes" (untuk yang negatif)

```python
df_2['class'].value_counts()
```

```python
# Jawaban anda di sini
df_2.loc[:,'class'] = np.where(df_2['class'].str.contains('neg'), 'Not diabetes',
                               np.where(df_2['class'].str.contains('Negative'), 'Not diabetes',
                                        np.where(df_2['class'].str.contains('pos'), 'Diabetes',
                                                np.where(df_2['class'].str.contains('Positive'), 'Diabetes','Other'))))
df_2
```

3. Dari jawaban nomor 1, ambil 3 kolom pertama yang berisi _missing value_ untuk diisi dengan _back filling_.

```python
# Jawaban anda di sini
df_2.iloc[:,:3] = df_2.iloc[:,:3].fillna(method='bfill')
df_2
```

4. Dari jawaban 1 dan 3, ambil 1 kolom selanjutnya yang berisi _missing value_ untuk diisi dengan nilai mediannya.

```python
# Jawaban anda di sini
col_med = df_2.iloc[:,3].median()
df_2.iloc[:,3].fillna(col_med, inplace=True)
df_2
```

5. Dari jawaban 1, 3, dan 4, ambil kolom sisanya yang berisi _missing value_ untuk diganti dengan nilai _mean_

```python
df_2.isna().sum()
```

```python
# Jawaban anda di sini
col_mean = df_2['body_mass_index'].mean()
df_2['body_mass_index'].fillna(col_mean, inplace=True)
```

```python
df_2.isna().sum()
```

6. Dari keseluruhan kolom numerik, ambil 4 kolom pertama untuk dinormalisasi

```python
df_2.dtypes
```

```python
# Jawaban anda di sini
from sklearn.preprocessing import normalize

norm_df = normalize(df_2.iloc[:,:4], norm='l2', axis=1)
norm_df
```

```python
df_norm = pd.DataFrame(norm_df, columns=df_2.columns[:4])
df_norm
```

```python
df_norm.isna().sum()
```

7. Dari soal nomor 6, ambil kolom sisanya (selain 4 kolom pertama) yang numerik untuk di-_scaling_ dengan membuat _range_-nya dari 0 hingga 1

```python
# Jawaban anda di sini
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler().fit_transform(df_2.iloc[:,4:7])
minmax
```

```python
df_minmax = pd.DataFrame(minmax, columns=df_2.columns[4:7])
df_minmax
```

```python
df_norm.merge(df_minmax,left_index=True,right_index=True)
df_norm
```
