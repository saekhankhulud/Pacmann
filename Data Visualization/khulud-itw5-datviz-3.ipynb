```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

Semua tugas menggunakan data yang sama dengan subbab 6

```python
df_train = pd.read_csv('data/train.csv')
df_train.describe().T
```

# Soal 1

Buat Scatter plot antara luas basement `TotalBsmtSF` dan luas lanatai 1 `1stFlrSF`. Tambahkan juga label pada masing masing axis dan judul.

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_figwidth(10)
fig.set_figheight(10)

ax.scatter(df_train['TotalBsmtSF'], df_train['1stFlrSF'],marker='*')
ax.grid()

ax.set_title('Perbandingan antara luas basement dan luas lantai 1')
ax.set_xlabel('Luas Area Basement (ft^2)')
ax.set_ylabel('Luas Area Lantai 1 (ft^2)')
```

# Soal 2

Tambahkan Kualitas rumah `OverallCond` pada plot sebelumnya. Tambahkan juga legendanya untuk tiap-tiap kualitas rumah. Lalu set skala pada sumbu Y menjadi nilai `log`.

```python
house_conditions = df_train['OverallCond'].value_counts().sort_values().index

fig, ax = plt.subplots(nrows=1, ncols=1)
fig.set_figwidth(10)
fig.set_figheight(10)

for house_condition in house_conditions:
    df_house_condition = df_train[df_train['OverallCond']==house_condition]
    ax.scatter(df_house_condition['TotalBsmtSF'],df_house_condition['1stFlrSF'],
              marker='x',label=house_condition)
ax.grid()

ax.set_title('Perbandingan antara luas basement dan luas lantai 1')
ax.set_xlabel('Luas Area Basement (ft^2)')
ax.set_ylabel('Luas Area Lantai 1 (ft^2)')
ax.set_yscale('log')
ax.legend()
```

# Soal 3

Dari soal sebelumnya, jadikan tiap-tiap kualitas sebagai subplots.

```python
house_conditions = df_train['OverallCond'].value_counts().sort_values().index
print('Jumlah kualitas rumah', len(house_conditions))
total_rows = 3
total_columns = 3

fig, ax = plt.subplots(nrows=3, ncols=3)
fig.set_figwidth(15)
fig.set_figheight(20)
fig.suptitle('Perbandingan antara luas basement dan luas lantai 1', fontsize=20)

for [num, house_condition] in enumerate(house_conditions):
    rows = num // total_rows
    column = num % total_rows
    df_house_condition = df_train[df_train['OverallCond']==house_condition]
    ax[column,rows].scatter(df_house_condition['TotalBsmtSF'], df_house_condition['1stFlrSF'],
                            marker='o',label=house_condition)
    ax[column,rows].grid()
    ax[column,rows].set_xlabel('Luas Area Basement (ft^2)')
    ax[column,rows].set_ylabel('Luas Area Lantai 1 (ft^2)')
    ax[column,rows].set_yscale('log')
    ax[column,rows].legend()
```

# Soal 4

Dari soal sebelumnya, jadikan subplots mempunyai axis x yang sama dan y yang sama.

```python
house_conditions = list(df_train['OverallCond'].value_counts().sort_values().index)
house_conditions.sort()
total_rows = 3
total_columns = 3

fig, ax = plt.subplots(nrows=3, ncols=3, sharey=True, sharex=True)
fig.subplots_adjust(hspace=0, wspace=0)
fig.set_figwidth(18)
fig.set_figheight(18)
fig.suptitle('Perbandingan antara luas basement dan luas lantai 1', fontsize=20)

for [num, house_condition] in enumerate(house_conditions):
    rows = num // total_rows
    column = num % total_rows
    df_house_condition = df_train[df_train['OverallCond']==house_condition]
    ax[rows][column].scatter(df_house_condition['TotalBsmtSF'], df_house_condition['1stFlrSF'], marker='o', label=house_condition)
    ax[rows][column].grid()
    
    ax[rows][column].set_yscale('log')
    
    if rows==total_rows-1:
        ax[rows][column].set_xlabel('Luas Area Basement (ft^2)')
        
    if column==0:
        ax[rows][column].set_ylabel('Luas Area Lantai 1 (ft^2)')
        
    ax[rows][column].set_ylim([100, 10000])
    ax[rows][column].set_xlim([-200, 6000])
    ax[rows][column].legend()
```

# Soal 5

Menggunakan subplots sebelumnya, tambahakan log harga rumah sebagai warna pada subplots yang masing masing colorbar ada pada sebelah axis.

```python
house_conditions = list(df_train['OverallCond'].value_counts().sort_values().index)
house_conditions.sort()
total_rows = 3
total_columns = 3

fig, ax = plt.subplots(nrows=3, ncols=3, sharey=True, sharex=True)
fig.subplots_adjust(hspace=0, wspace=0)
fig.set_figwidth(18)
fig.set_figheight(18)
fig.suptitle('Perbandingan antara luas basement dan luas lantai 1', fontsize=20)

for [num, house_condition] in enumerate(house_conditions):
    rows = num // total_rows
    column = num % total_rows
    df_house_condition = df_train[df_train['OverallCond']==house_condition]
    mainPlot = ax[rows][column].scatter(
        df_house_condition['TotalBsmtSF'],
        df_house_condition['1stFlrSF'],
        c = df_house_condition['SalePrice'],
        marker='o',
        label=house_condition
    )
    ax[rows][column].grid()
    
    ax[rows][column].set_yscale('log')
    
    if rows==total_rows-1:
        ax[rows][column].set_xlabel('Luas Area Basement (ft^2)')
        
    if column==0:
        ax[rows][column].set_ylabel('Luas Area Lantai 1 (ft^2)')
        
    ax[rows][column].set_ylim([100, 10000])
    ax[rows][column].set_xlim([-200, 6000])
    ax[rows][column].legend()
    fig.colorbar(mainPlot, ax=ax[rows][column])
```

# Soal 6

Menggunakan subplots sebelumnya, tambahakan log harga rumah sebagai satu colorbar, dan semua harga pada setiap subplots merujuk ke satu colorbar tersebut.

```python
from matplotlib.cm import ScalarMappable

house_conditions = list(df_train['OverallCond'].value_counts().sort_values().index)
house_conditions.sort()
total_rows = 3
total_columns = 3
df_train['LogSalePrice'] = np.log(df_train['SalePrice'])

cmap = plt.get_cmap("viridis")
norm = plt.Normalize(df_train['LogSalePrice'].min(), df_train['LogSalePrice'].max())

fig, ax = plt.subplots(nrows=3, ncols=3, sharey=True, sharex=True)
fig.subplots_adjust(hspace=0, wspace=0)
fig.set_figwidth(18)
fig.set_figheight(18)
fig.suptitle('Perbandingan antara luas basement dan luas lantai 1', fontsize=20)

for [num, house_condition] in enumerate(house_conditions):
    rows = num // total_rows
    column = num % total_rows
    ax[rows][column].text(0.5, 0.5, house_condition, transform= ax[rows][column].transAxes,
        fontsize=100, color='gray', alpha=0.3,
        ha='center', va='center')
    df_house_condition = df_train[df_train['OverallCond']==house_condition]
    mainPlot = ax[rows][column].scatter(
        df_house_condition['TotalBsmtSF'],
        df_house_condition['1stFlrSF'],
        c = cmap(norm(list(df_house_condition['LogSalePrice']))),
        marker='o',
        label=house_condition,
        alpha=.4
    )
    ax[rows][column].grid()
    
    ax[rows][column].set_yscale('log')
    
    if rows==total_rows-1:
        ax[rows][column].set_xlabel('Luas Area Basement (ft^2)')
        
    if column==0:
        ax[rows][column].set_ylabel('Luas Area Lantai 1 (ft^2)')
        
    ax[rows][column].set_ylim([100, 10000])
    ax[rows][column].set_xlim([-200, 6000])
    
sm = ScalarMappable(norm=norm, cmap=cmap)
sm.set_array([])
cax = fig.add_axes([0.15, .92, 0.7, 0.01])
cbar = fig.colorbar(sm, ax=ax, orientation='horizontal', cax=cax)
cbar.ax.set_title("Harga Rumah (Log Scale)")
```
